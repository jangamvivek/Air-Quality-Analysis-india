```python
df = pd.read_csv('air_pollution_data.csv')
df
```

```python
pollutants = ['aqi', 'no', 'no2', 'o3', 'so2', 'pm2_5', 'pm10', 'nh3']

# Find cities and dates with the highest and lowest values for each pollutant with date
city_stats_with_dates = {}

for pollutant in pollutants:
    highest = df.loc[df[pollutant].idxmax()]  # Row with max value
    lowest = df.loc[df[pollutant].idxmin()]   # Row with min value
    
    city_stats_with_dates[pollutant] = {
        'Highest': {
            'City': highest['city'],
            'Date': highest['date'],
            'Value': highest[pollutant]
        },
        'Lowest': {
            'City': lowest['city'],
            'Date': lowest['date'],
            'Value': lowest[pollutant]
        }
    }

# Display results
for pollutant, stats in city_stats_with_dates.items():
    print(f"{pollutant}:")
    print(f"  Highest: {stats['Highest']['City']} on {stats['Highest']['Date']} with value {stats['Highest']['Value']}")
    print(f"  Lowest: {stats['Lowest']['City']} on {stats['Lowest']['Date']} with value {stats['Lowest']['Value']}")
```

```python
# Prepare results
pollutant_summary = []
for pollutant, stats in city_stats_with_dates.items():
    pollutant_summary.append({
        'Pollutant': pollutant,
        'Highest City': stats['Highest']['City'],
        'Highest Date': stats['Highest']['Date'],
        'Highest Value': stats['Highest']['Value'],
        'Lowest City': stats['Lowest']['City'],
        'Lowest Date': stats['Lowest']['Date'],
        'Lowest Value': stats['Lowest']['Value']
    })

# Convert to DataFrame for better presentation
summary_df = pd.DataFrame(pollutant_summary)

# Save to CSV (optional, for external use)
summary_df.to_csv('Pollutant_City_Report.csv', index=False)

print(summary_df)
```

```python
from reportlab.lib.pagesizes import letter
from reportlab.platypus import SimpleDocTemplate, Table, TableStyle
from reportlab.lib import colors

# Create PDF
doc = SimpleDocTemplate("Air_Quality_Report.pdf", pagesize=letter)

# Table data
table_data = [['Pollutant', 'Highest City', 'Highest Date', 'Highest Value',
               'Lowest City', 'Lowest Date', 'Lowest Value']] + summary_df.values.tolist()

# Create table
table = Table(table_data)
table.setStyle(TableStyle([
    ('BACKGROUND', (0, 0), (-1, 0), colors.grey),
    ('TEXTCOLOR', (0, 0), (-1, 0), colors.whitesmoke),
    ('ALIGN', (0, 0), (-1, -1), 'CENTER'),
    ('FONTNAME', (0, 0), (-1, 0), 'Helvetica-Bold'),
    ('BOTTOMPADDING', (0, 0), (-1, 0), 12),
    ('BACKGROUND', (0, 1), (-1, -1), colors.beige),
    ('GRID', (0, 0), (-1, -1), 1, colors.black)
]))

# Build PDF
doc.build([table])
```
